###We can filter the data by providing some condtions using loc() function

```python
df.loc[df['Type 1'] == 'Fire']  # This will print all the rows for which the 'Type 1' is Fire
```

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]  # This will show only those rows where Type 1 will be Grass and Type 2 will be poison only
```

```python
df.loc[(df['Type 1'] == 'Water') | (df['Type 2'] == 'Rock')]  # This will show those rows where Type 1 will be Water or Type 2 will be Rock.
```

```python
df.loc[df['HP'] > 100]  # This will print all those rows where 'HP' is greater than 100
```

```python
# We will filter out all the names containing the word 'Mega'
df.loc[df['Name'].str.contains('Mega')]
```

```python
# We will filter out all those names not containing the word 'Mega' by using the name '~'
df.loc[~df['Name'].str.contains('Mega')]
```

```python
# This will filter all the names starting from 'Pi' using regular expression
df.loc[df['Name'].str.contains('^Pi[a-z]*', regex=True)]
```

```python
# Change a particular word in the dataframe like whereever the word 'Nan' is written in Type 1 we want it to be 'Solid Rock'
df.loc[df['Type 1'] == 'Nan', 'Type 1'] = 'Solid Rock'
df
```

###Sorting the data

```python
df.sort_values('Name', ascending=False)  # This will sort all names in Names column but in reverse alphabetical order
```

```python
df.sort_values(['HP', 'Attack'], ascending=[1, 0])  # This will sort HP column in asc and Attack column in desc(1 means ascending & 0 means descending)
```

###Making changes to the data

```python
df.head(10)
```

```python
# Create a new column in the dataset called "Total" which includes the sum of some columns
df['Total Power'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

```python
# We can also sum the columns like this
# We will create another column
df['Total Sum'] = df.iloc[:, 3:10].sum(axis=1)  # In [:, 3:10], first ':' means all the rows and '3:10' means from 3rd column to 9 column
df.head(5)
```

```python
# Now remove the extra Total column that we have created just to show the another way of doing sum of columns
df = df.drop(columns=['Total Sum'])
```

```python
df.head(5)
```

###Re-arranging or Organizing the column headers

```python
df = df[['Name', 'Type 1', 'Type 2', 'Legendary', 'HP', 'Attack', 'Defense', 'Speed', 'Sp. Atk', 'Sp. Def', 'Total Power']]
df
```

######As Typing all header names can be a bit annoying, so we can also re-arrange the column headers by its index position and at once

##Aggregate Statistics (GroupBy)

######Suppose we want to see the avg HP or Attack of all the pokemon groupby for the type they are

```python
df.groupby(['Type 1']).mean()  # averages of all the Type 1 pokemon
```

```python
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)  # sorting the average of defense
# steel type pokemon has the highest defense
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)  # sorting the average of average
# dragon has the highest attack
```

```python
df.groupby(['Type 1']).sum()
```

```python
df.groupby(['Type 1']).count()
```

```python
df['count'] = 1  # creating column named count
df.groupby(['Type 1']).count()['count']
```

```python
df
```

##Working with large amount of data

```python
new_df = pd.DataFrame(columns=df.columns)  # creating new dataframe with the same column as orginal dataframe

for dataf in pd.read_csv('modified.csv', chunksize=5):   # means 5 rows will get passed at a time, 'dataf' contains only 5 rows of the dataset, this will load the dataset but in chunk of 5

  result = df.groupby(['Type 1']).count()  # getting count of Type 1 column
  new_df = pd.concat([new_df, result])  # concatenting the two dataframes i.e, df and new_df, basically it takes your new dataframe as you through chunks, appends on results and store it back to new dataframe
```

```python
new_df
```
